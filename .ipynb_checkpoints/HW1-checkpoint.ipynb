```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('post-operative.csv', header=None, names=['L_CORE', 'L_SURF', 'L_O2', 'L_BP', 'SURF_STBL', 'CORE_STBL', 'BP_STBL', 'COMFORT', 'ADM_DECS'])
```

```python
df
```

```python
compare = []
for header in list(df):
    if(header != 'COMFORT'):
        compare.append(header)


        
g = sns.PairGrid(df,
                 x_vars=compare,
                 y_vars=["COMFORT"],
                 aspect=.75, size=4)
g.map(sns.violinplot, palette="pastel");
```

```python
# Why is there an extra A in the ADM_DECS plot?

set(df.ADM_DECS)
```

```python
# As we can see, one of the 'A' elements consists of an extra space. Let's get rid of it.

df.ADM_DECS = df.ADM_DECS.replace(['A '], 'A')
```

```python
# Now that we fixed the issue, let's replot it!


g = sns.PairGrid(df,
                 x_vars=compare,
                 y_vars=["COMFORT"],
                 aspect=.75, size=4)
g.map(sns.violinplot, palette="pastel");
```

# Preprocessing Data

### L_CORE, L_SURF, L_BP

All three of these headers have values that are represented using 'low', 'mid', and 'high'. Let's change them to 1, 2, and 3 respectively

```python
df.replace(['low'],1,inplace=True)
df.replace(['mid'],2,inplace=True)
df.replace(['high'],3,inplace=True)
```

### L_O2

I'm changing 'good' to 1 and 'excellent' to 2

```python
df.L_O2.replace(['good'],1, inplace=True)
df.L_O2.replace(['excellent'],2, inplace=True)
```

### SURF-STBL, CORE-STBL, BP-STBL

We can change 'unstable', 'mod-stable', and 'stable' to -1, 0, and 1 respectively.

```python
df.replace(['unstable'],-1,inplace=True)
df.replace(['mod-stable'],0,inplace=True)
df.replace(['stable'],1,inplace=True)
```

### ADM-DECS

I'm going to change I, S, A to 1, 2, and 3 respectively

```python
df.ADM_DECS.replace(['I'],1, inplace=True)
df.ADM_DECS.replace(['S'],2, inplace=True)
df.ADM_DECS.replace(['A'],3, inplace=True)
```
